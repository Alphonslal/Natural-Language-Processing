```python
#process 1
import pandas as pd

data = {
    "text": [
        "I absolutely love this new phone â€” the camera is fantastic",
        "The room service was terrible and Iâ€™m very disappointed with my stay",
        "Itâ€™s okay, the product works, but itâ€™s not quite what I expected",
        "I canâ€™t believe how quickly they fixed my issue â€” excellent customer support.",
        "While the design is beautiful, the battery life is awful and drains too fast"
    ]
}

df = pd.DataFrame(data)
df
```

```python
#Process 2
import nltk
nltk.download('vader_lexicon')

from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

df['score'] = df['text'].apply(lambda x: sia.polarity_scores(x)['compound'])
df['sentiment'] = df['score'].apply(lambda x:
                                    'Positive' if x > 0
                                    else ('Negative' if x < 0
                                    else 'Neutral'))
df
```

Converting text to lowercase in sentiment analysis (and more broadly in NLP) is a common preprocessing step for several good reasons:

It ensures uniformity of tokens by treating words that differ only by capitalization (e.g., â€œAppleâ€ vs. â€œappleâ€) as the same word, thereby reducing duplication in vocabulary.

```python
import string
df['no_punct'] = df['clean_text'].str.translate(str.maketrans('', '', string.punctuation))
df[['clean_text','no_punct']]
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop = set(stopwords.words('english'))
df['no_stopwords'] = df['no_punct'].apply(lambda x: " ".join([word for word in x.split() if word not in stop]))
df[['no_punct','no_stopwords']]
```

```python
import re

def remove_emojis(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"
                           u"\U0001F300-\U0001F5FF"
                           u"\U0001F680-\U0001F6FF"
                           u"\U0001F1E0-\U0001F1FF"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)

df['no_emoji'] = df['no_stopwords'].apply(remove_emojis)
df[['no_stopwords','no_emoji']]
```
